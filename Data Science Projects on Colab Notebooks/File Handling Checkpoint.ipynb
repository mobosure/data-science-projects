```python
#Importing the numpy library and aliasing it as np

import numpy as np
```

```python
#Using the open() function to open the loan dataset csv file and assigning the result to a variable called checkpoint_file.

checkpoint_file = np.genfromtxt(r'Loan_prediction_dataset.csv',  delimiter = ',', dtype = str  )
```

```python
checkpoint_file
```

```python
#Checking to know the columns available

checkpoint_file[0]
```

```python
#Indexing out the Loan Amount column that we want to calculate the mean, median, and standard deviation on

checkpoint_file[1:,8]
```

```python
checkpoint_file[1:,8]
```

```python
#Using for loop to convert rows containing '' to 0 in the Loan Amount Column

for i in range(len(checkpoint_file[1:,8])):
  if checkpoint_file[1:,8][i] == '':
    checkpoint_file[1:,8][i] = 0
  else:
    pass
```

```python
checkpoint_file[1:,8]
```

```python
#Assigning the rows on the loan amount column to a new variable called loan_amount_column and coverting the datatype from unicode to float

loan_amount_column = checkpoint_file[1:,8].astype('float64')
```

```python
loan_amount_column.dtype
```

```python
# Using the numpy array to find the mean of the loan amounts.

np.mean(loan_amount_column)
```

```python
# Using the numpy array to find the median of the loan amounts.

np.median(loan_amount_column)
```

```python
# Using the numpy array to find the standard deviation of the loan amounts.

np.std(loan_amount_column)
```
