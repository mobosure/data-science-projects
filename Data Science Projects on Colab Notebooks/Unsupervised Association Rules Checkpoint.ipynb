```python
import warnings
warnings.filterwarnings('ignore')
import mlxtend
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
import pandas as pd
import numpy as np
```

```python
import csv
files = '/content/Market_Basket_Optimisation.csv'
with open(files) as f:
  reader = csv.reader(f)
  data = list(reader)

data
```

```python
encoder = TransactionEncoder() # ----------------------------------------------- instantiate the encoder
encoded = encoder.fit(data).transform(data) # -------------------------------- Fit transform the new data (list of lists)
encoded

#  Turn encoded data to a dataframe. 'True' to be 1, and all 'False' to be 0
encodedData = pd.DataFrame(data = encoded, columns = encoder.columns_, dtype = int)
encodedData
```

```python
support = apriori(encodedData, min_support= 0.05,use_colnames = True)
support.sort_values(by = 'support', ascending = False)

confidence = association_rules(support, metric = 'confidence', min_threshold = 0.1)
confidence.sort_values(by = 'confidence', ascending = False)
```
