```python
import numpy as np
```

```python
# creating the "grades" array from the grades list

grades = [85, 90, 88, 92, 95, 80, 75, 98, 89, 83]

grades = np.array(grades)
```

```python
# Using numpy functions to calculate the mean of the grades
np.mean(grades)
```

```python
# Using numpy functions to calculate the median of the grades

np.median(grades)
```

```python
# Using numpy functions to calculate the standard deviation of the grades

np.std(grades)
```

```python
# Using numpy function to find the maximum of the grades

np.max(grades)
```

```python
# Using numpy function to find the minimum of the grades

np.min(grades)
```

```python
# Using numpy function to sort the grades in ascending order.

np.sort(grades)
```

```python
# Using numpy function to find the index of the highest grade in the array.

np.argmax(grades)
```

```python
# Using numpy function to count the number of students who scored above 90.

np.count_nonzero(grades>90)
```

```python
# Using numpy function to calculate the percentage of students who scored above 90.

np.mean(grades > 90) * 100
```

```python
# Using numpy function to calculate the percentage of students who scored below 75.

np.mean(grades < 75) * 100
```

```python
# Using numpy function to extract all the grades above 90 and put them in a new array called "high_performers".

high_performers = grades[grades>90]
high_performers
```

```python
# Creating a new array called "passing_grades" that contains all the grades above 75.

passing_grades = grades[grades>75]
passing_grades
```
