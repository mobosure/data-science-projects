```python
# Creating a class called "Account"
class Account:
  # Creating the following attributes of the Account class: account_number (string), account_balance (float) and account_holder (string)
  def __init__(self, account_number, account_balance, account_holder):
    self.account_number = str(account_number)
    self.account_balance = float(account_balance)
    self.account_holder = str(account_holder)

  # Creating a method called deposit that adds the amount passed as an argument to the account_balance
  def deposit(self, amount):
      self.account_balance += amount
      return self.account_balance

  # Creating a method called withdraw that subtracts the amount passed as an argument to the account_balance
  def withdraw(self, amount):
      if self.account_balance >= amount:
          self.account_balance -= amount
          return self.account_balance

  # Creating a method called check_balance that return the current account balance
  def check_balance(self):
      return self.account_balance

# Testing the program by creating multiple instances of the Account class and performing different transactions on them.
person1 = Account(556789005, 5000, 'Madago Emuobo Prosper')
print(person1.account_number)
print(person1.account_balance)
print(person1.account_holder)
person1.deposit(1000)
person1.withdraw(500)
person1.check_balance()

person2 = Account(5567890009, 43000, 'Mr Smith')
print(person2.account_number)
print(person2.account_balance)
print(person2.account_holder)
person2.deposit(41000)
person2.withdraw(4300)
person2.check_balance()
```
